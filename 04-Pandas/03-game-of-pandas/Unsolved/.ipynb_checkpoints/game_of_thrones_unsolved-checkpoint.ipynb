```python
# Import Dependencies
import pandas as pd
```

```python
# Save file path to variable
file = "../Resources/game_of_thrones.csv"
```

```python
# Read csv with Pandas
thrones_data = pd.read_csv(file)
thrones_data.head()
```

```python
# Collect a list of all the unique values in "Allegiances"
allegiances = thrones_data["Allegiances"].unique()
allegiances
```

```python
# Create a new table with only the Allegiances and Names
sub_data = thrones_data[["Name", "Allegiances"]]
sub_data.head()
```

```python
# Create a GroupBy object based on the "Allegiances" column in the newley created table
group = thrones_data.groupby("Allegiances")

# Count the amount of deaths that has occured in each house
# Rename the "Name" column to "Death Count"
death_count = pd.DataFrame(group.count()["Death Year"])
death_count = death_count.rename(columns = {"Death Year": "Death Count"})
death_count
```

```python
# Sort by most deaths and reset index
death_count = death_count.sort_values("Death Count", ascending = False)
death_count = death_count.reset_index()
death_count
```

```python
# Save the death count for the Night's Watch, using `iloc` to filter
nw_death_count = death_count.iloc[1, 1]
nw_death_count
```
