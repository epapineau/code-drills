```python
# Import numpy, matplotlib, and pandas
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Load the csv from Resources folder as school_ranking_df
school_ranking_df = pd.read_csv("./Resources/cwurData.csv")

# Show the first 5 rows of school_ranking_df
school_ranking_df.head()
```

```python
# Declare a variable called institution_avg_score_df 
    # Set it with the group of institution and average of score
    # Hint: When grouping, use as_index=False to retain column names.
institution_avg_score_df = school_ranking_df.groupby("institution", as_index = False)
institution_avg_score_df = institution_avg_score_df["score"].mean()

# Show the first 10 rows of school_ranking_df
institution_avg_score_df.head(10)
```

```python
# Declare a variable called top_avg_sorted_df 
    # Sort institution_avg_score_df by score from Highest to Lowest.
    # then save only the top 10 rows
top_avg_sorted_df = institution_avg_score_df.sort_values("score", ascending = False)
top_avg_sorted_df = top_avg_sorted_df[0:9]

# Show the rows of top_avg_sorted_df
top_avg_sorted_df
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(top_avg_sorted_df["institution"]))
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
# Bars should indicate score
# Horizontal axis should show the institutions
# Set x axis and tick locations
plt.figure(figsize =(10, 8))
plt.bar(x_axis, top_avg_sorted_df["score"], color = "r", alpha = .5)
plt.xticks(x_axis, top_avg_sorted_df["institution"], rotation = 90)
```

```python
# Set x and y limits
plt.xlim(-.5, max(x_axis) + .5)
```

```python
# Set a title and labels
plt.title("Average Score of Top 10 Institutions 2012 - 2015")
plt.xlabel("Institutions")
plt.ylabel("Average Score (%)")
```

```python
# Save our graph and show the grap
plt.tight_layout()
plt.savefig("./images/drill02.png")
plt.show
```
