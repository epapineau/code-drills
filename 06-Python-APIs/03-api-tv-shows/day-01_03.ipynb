```python
# Import your dependencies requests, json, and pprint for pretty printing.
import requests
import json
import pprint as pp
import numpy as np
```

```python
# Create a variable called search_tv
    # Request for user input with the question "What show would you like?"
        # Example "friends" as an input
search_tv = input("What show would you like? ")
```

```python
# Create a variable called base_url and 
    # set it to the string below
    # "http://api.tvmaze.com/search/shows?q="
base_url = "http://api.tvmaze.com/search/shows?q="
```

```python
# Set the variable 'response' to the data from base_url with search_tv appended
response = requests.get(f"{base_url}{search_tv}")
```

```python
# Create a variable called res_data 
    # Set it to the response converted as a json format
res_data = response.json()

# Using pprint, print the res_data
pp.pprint(res_data)
```

```python
# Print the following keys from the response JSON.
    # "Show name:" Name
    # "Show score:" Score
    # "Show rating:" Rating
print(f"Show name: {res_data[0]['show']['name']}")
print(f"Show score: {res_data[0]['score']}")
print(f"Show rating: {res_data[0]['show']['rating']['average']}")
```

```python
# Loop through each item from the response and print the show's name and rating.
    # "Show name: " Name
    # "Show rating:" Rating

```

```python
len(res_data)
for i in np.arange(len(res_data)):
    print(f"Show name: {res_data[i]['show']['name']}")
    print(f"Show rating: {res_data[i]['show']['rating']['average']}")
```
