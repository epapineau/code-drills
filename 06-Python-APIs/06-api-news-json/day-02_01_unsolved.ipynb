```python
# Dependencies json, requests, pprint, and your config file for api_key
import json
import requests
import pprint as pp
from config import api_key
```

```python
# Set the url and query information for the country of US
base_url = f"https://newsapi.org//v2/top-headlines?apiKey={api_key}&country="
country = 'US'

# Build query URL with the country query.
q_url = f"{base_url}{country}"
```

```python
# Get news data using requests
news_json = requests.get(q_url).json()

# Test your api call using by printing your api data.
print(f"The news API responded with: {news_json}.")
```

```python
# Create variable called articles_list 
    # Set it to a list of all article dictionaries from news_json
articles_list = news_json["articles"]

# Use pprint to print the articles_list
pp.pprint(articles_list)
```
