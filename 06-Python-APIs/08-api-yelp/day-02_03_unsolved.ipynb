```python
# Dependencies numpy as 'np', requests, matplotlib's pyplot as "plt", pandas as "pd", pprint as pprint,
# and your config file for api_key
import json
import requests
import pprint as pp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from config import api_key
```

```python
# Set the base URL for Yelp and parameters for category and location
    # Query for the following information:
    # url is "https://api.yelp.com/v3/businesses/search?"
        # Please review documentation to get the base url for searching restaurants.
            # https://www.yelp.com/developers/documentation/v3/business_search
        # If still stuck, check 'hint.txt' for base url
    # category is "chinese"
    # location is "san francisco" 
    # result limit is 50
url = "https://api.yelp.com/v3/businesses/search?"
category = "chinese"
location = "san francisco"
lim = 50

# Create the query URL with the parameters
# hint - remember that parameters in a given query url are separated by "&"
q_url = f"{url}location={location}&categories={category}&limit={lim}"

# headers are a bearer token that is required by Yelp's API. We have provided the syntax for you
headers = {"authorization" : "Bearer " + api_key}
```

```python
# Call the API and store the response as a JSON, we're providing you the example with headers.
    # However, be sure to review the documentation for requests using headers
    # http://docs.python-requests.org/en/master/user/quickstart/
response = requests.get(q_url, headers = headers).json()

# Use pprint to print the JSON object
pp.pprint(response)
```

```python
# Create the following variables and set them to empty lists
    # zipcode_list, rating_list
zipcode_list = []
rating_list = []

# Iterate through the response and append the zip codes to zipcode_list and ratings to rating_list
for i in np.arange(len(response['businesses'])):
    zipcode_list.append(response['businesses'][i]['location']['zip_code'])
    rating_list.append(response['businesses'][i]['rating'])
    
# print both lists to check if you've grabbed the correct data
print(zipcode_list)
print(rating_list)
```

```python
# Create a dictionary that contains zip code and rating
yelp_dict = {'zip_code': zipcode_list,
            'rating': rating_list}

# Declare a variable called "yelp_data" and convert the dictionary to a panda dataframe
yelp_data = pd.DataFrame(yelp_dict)

# Display the first 5 rows of yelp_data
yelp_data.head()
```

```python
# Declare a variable called 'zip_group' and find the average rating for each zip code from the yelp_data dataframe.
    # Hint: when grouping, you'll want to keep your zip code as a column not index. 
    # Check out 'as_index' in documentation for groupby.
zip_group = yelp_data.groupby("zip_code", as_index = False).mean()
zip_group
```

```python
# Create an x-axis based on the length of zip_group's ratings
x_axis = np.arange(len(zip_group))
x_axis
```

```python
# Plot the figure with a width of 15 and height of 3
fig, ax = plt.subplots(figsize = (15, 3))
ax.bar(x_axis,
       zip_group['rating'],
       color='r')

# Add labels to each x tick from the zip_group's zip_code column. 
ax.set_xticks(x_axis)
ax.set_xticklabels(zip_group['zip_code'])

# Add titles and labels to graph.
plt.title("Average Restaurant Rating by Zip Code")
plt.ylabel("Average Restaurant Rating")
plt.xlabel("Zip Code")

#Plot the data in a bar graph based on the ratings
# Apply tight layout to remove extra white space and save the graph to images folder.
plt.tight_layout()
plt.show()
```
